```python
lookup = {
        'one': 1,
         'two': 2,
         'three': 3,
         'four': 4,
         'five': 5,
         'six': 6,
         'seven': 7,
         'eight': 8,
         'nine': 9,}

lookup_forward = {}

for k in list(lookup.keys())[::-1]:
    for i in lookup.keys():
        if k[-1] == i[0]:
            lookup_forward[k+i[1:]] = lookup[k]
```

```python
lookup_forward.update(lookup)
```

```python
lookup_backward_base = {}
for k in lookup.keys():
    lookup_backward_base[k[::-1]] = lookup[k] 
```

```python
lookup_backward = {}
for k in list(lookup_backward_base.keys())[::-1]:
    for i in lookup_backward_base.keys():
        if k[0] == i[-1]:
            lookup_backward[i+k[1:]] = lookup_backward_base[k]
```

```python
lookup_backward.update(lookup_backward_base)
```

```python
with open('data/day1.txt', 'r') as f:
    lf = f.read()
for i in lookup_forward.keys():
    lf = lf.replace(i, str(lookup_forward[i]))
    
lf = lf.split('\n')
```

```python
with open('data/day1.txt', 'r') as f:
    lb = f.read()
for i in lookup_backward.keys():
    lb = lb[::-1].replace(i, str(lookup_backward[i]))[::-1]
lb = lb.split('\n')
```
